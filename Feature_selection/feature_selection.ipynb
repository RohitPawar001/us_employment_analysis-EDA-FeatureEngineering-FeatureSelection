```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.pandas.set_option("display.max_columns",None)
```

```python
data = pd.read_csv("/content/us_employment_fe.csv")
```

```python
data.head()
```

```python
data.drop(["Unnamed: 0.1"],axis = 1,inplace = True)
```

```python
data.drop(["Unnamed: 0"],axis = 1,inplace = True)
```

```python
x = data[['year', 'population', 'labor_force', 'population_percent',
       'employed_percent', 'agrictulture_ratio',
       'nonagriculture_ratio', 'unemployed', 'unemployed_percent',
       'not_in_labor']]
y = data['employed_total']
```

this dataset can be used for solve the regression type of problem's so we are using chi2 for model selection.

##**CHI SQUARE TEST**

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
ordered_rank_featurre = SelectKBest(score_func= chi2,k =10)
```

```python
ordered_rank = ordered_rank_featurre.fit(x,y)
```

```python
ordered_rank.scores_
```

```python
df = pd.DataFrame(ordered_rank.scores_)
```

```python
df
```

```python
df_columns = pd.DataFrame(x.columns)
```

```python
selected_feature = pd.concat([df_columns,df],axis=1)
```

```python
selected_feature
```

```python
selected_feature.columns = ['feature','score']
```

```python
selected_feature
```

```python
selected_feature.sort_values(by="score",ascending=False)
```

from the above observation we can say that the population, non agriculture and labour force has a higher priority than others.
